```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

# Part 1 - MNIST FCNN

```python
(x_train_mnist, y_train_mnist), (x_test_mnist, y_test_mnist) = tf.keras.datasets.mnist.load_data()


x_train_mnist = x_train_mnist / 255.0
x_test_mnist = x_test_mnist / 255.0


mnist_fcnn = tf.keras.Sequential([
    tf.keras.Input(shape=(28, 28)),
    tf.keras.layers.Flatten(name="flatten_28x28_to_784"),
    tf.keras.layers.Dense(128, activation="relu", name="hidden_dense_1"),
    tf.keras.layers.Dense(64, activation="relu", name="hidden_dense_2"),
    tf.keras.layers.Dense(10, activation="softmax", name="output_layer")
])

mnist_fcnn.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

mnist_fcnn.summary()

history_mnist = mnist_fcnn.fit(x_train_mnist, y_train_mnist, epochs=10, validation_split=0.2
)

mnist_test_loss, mnist_test_accuracy = mnist_fcnn.evaluate(x_test_mnist, y_test_mnist)

print("MNIST Test Loss:", mnist_test_loss)
print("MNIST Test Accuracy:", mnist_test_accuracy)
```

# Part 2 — CIFAR-10 FCNN Experiment

```python
(x_train_cifar, y_train_cifar), (x_test_cifar, y_test_cifar) = tf.keras.datasets.cifar10.load_data()

x_train_cifar = x_train_cifar / 255.0
x_test_cifar = x_test_cifar / 255.0

y_train_cifar = y_train_cifar.reshape(-1)
y_test_cifar = y_test_cifar.reshape(-1)

cifar_fcnn = tf.keras.Sequential([
    tf.keras.Input(shape=(32, 32, 3)),
    tf.keras.layers.Flatten(name="flatten_32x32x3_to_3072"),
    tf.keras.layers.Dense(128, activation="relu", name="hidden_dense_1"),
    tf.keras.layers.Dense(64, activation="relu", name="hidden_dense_2"),
    tf.keras.layers.Dense(10, activation="softmax", name="output_layer")
])

cifar_fcnn.compile(optimizer="adam", loss="sparse_categorical_crossentropy",)

cifar_fcnn.summary()

history_cifar = cifar_fcnn.fit(x_train_cifar, y_train_cifar, epochs=10, validation_split=0.2)

cifar_test_loss, cifar_test_accuracy = cifar_fcnn.evaluate(x_test_cifar, y_test_cifar)

print("CIFAR-10 Test Loss:", cifar_test_loss)
print("CIFAR-10 Test Accuracy:", cifar_test_accuracy)
```

# Part 3 — Compare Results

```python
print("MNIST FCNN Test Accuracy:", mnist_test_accuracy)
print("CIFAR-10 FCNN Test Accuracy:", cifar_test_accuracy)

print("MNIST FCNN Test Loss:", mnist_test_loss)
print("CIFAR-10 FCNN Test Loss:", cifar_test_loss)
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history_mnist.history["accuracy"], label="Training Accuracy")
plt.plot(history_mnist.history["val_accuracy"], label="Validation Accuracy")
plt.title("MNIST FCNN Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history_mnist.history["loss"], label="Training Loss")
plt.plot(history_mnist.history["val_loss"], label="Validation Loss")
plt.title("MNIST FCNN Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history_cifar.history["accuracy"], label="Training Accuracy")
plt.plot(history_cifar.history["val_accuracy"], label="Validation Accuracy")
plt.title("CIFAR-10 FCNN Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history_cifar.history["loss"], label="Training Loss")
plt.plot(history_cifar.history["val_loss"], label="Validation Loss")
plt.title("CIFAR-10 FCNN Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

The same basic FCNN idea performed very well on MNIST, but its performance dropped on CIFAR-10. This shows that FCNN can be effective for simple grayscale images, but it struggles more with RGB object images because the data has more inputs, more color information, more background variation, and more complex spatial structure.
